```python
import pandas as pd
import numpy as np
```

```python
usage = pd.read_csv('data/usage.csv')
# display(usage)
entry_counts = pd.read_csv('data/entry_counts.csv')
# display(entry_counts)
usage_contests = usage['contest_id'].unique()
# print(len(usage_contests))
usage_data = usage.merge(entry_counts, how='inner', on='contest_id')
usage_data['usage_ratio'] = usage_data['entry_count_x'] / usage_data['entry_count_y']
usage_data = usage_data.drop(columns=['entry_count_x']).rename(columns={'entry_count_y': 'entry_count'})
display(usage_data)
print(len(usage_data))
usage_check = usage_data.groupby('contest_id')[['usage_ratio']].sum().sort_values(by='usage_ratio', ascending=False)
# display(usage_check.value_counts())
```

```python
contests = pd.read_csv('data/contest_data.csv')
# display(contests)
usage_data_with_contests = usage_data.merge(contests, on=['contest_id'])
display(usage_data_with_contests)
print(usage_data_with_contests.columns)
```

```python
draftables = pd.read_csv('data/draftables_data.csv')
display(draftables[draftables['salary'].isna()])
usage_data_all = usage_data_with_contests.merge(draftables, on=['draft_group_id', 'player_id', 'roster_slot_id'])
usage_data_all = usage_data_all.rename(columns={'name_x': 'contest_name', 'name_y': 'player_name'})
display(usage_data_all)
usage_data_all['competition_id'] = usage_data_all['competition_id'].astype(int)
```

```python
# need to match up player names from projection data
import data_wrangler as dw
from fuzzywuzzy import process
import json

def match_player_name(player_name, potential_matches):
    print('Matching for player:')
    print(player_name)
    matches = process.extract(player_name, potential_matches, limit=10)
    print('Potential matches:')
    print(matches)
    print(f'"{player_name}": "{matches[0][0]}"')

projection_data = dw.get_filtered_projection_data()
display(projection_data)
dst_unique = usage_data_all[usage_data_all['position'] == 'DST'].groupby(['player_name', 'team_abbreviation']).groups.keys()
dk_def_map = {val[1]: val[0] for val in dst_unique}
print(dk_def_map)
projection_data.loc[projection_data['position'] == 'DST', 'player_name'] = projection_data.loc[projection_data['position'] == 'DST', 'player_name'].map(lambda x: dk_def_map.get(x))
with open('data/player_name_dict.json', 'r') as pdfile:
    player_dict = json.load(pdfile)
projection_data['player_name'] = projection_data['player_name'].map(lambda x: player_dict.get(x, x))


usage_data_with_projections = usage_data_all.merge(projection_data, on=['season', 'week', 'player_name', 'position'], how='left')
usage_data_with_projections['projection'] = usage_data_with_projections['projection'].fillna(0.0)

# fill in contest types
contest_type_dict = {
    96: 'Showdown',
    51: 'Tiers',
    21: 'Classic'
}

contest_types = ['Showdown', 'Classic', 'Tiers']
usage_data_with_projections['contest_type'] = usage_data_with_projections['contest_type'].map(lambda x: contest_type_dict.get(x, x))
# fill in roster slots
roster_slot_dict = {
    66: 'QB',
    67: 'RB',
    68: 'WR',
    69: 'TE',
    70: 'FLEX',
    71: 'DST',
    323: 'T1',
    324: 'T2',
    325: 'T3',
    326: 'T4',
    327: 'T5',
    328: 'T6',
    511: 'CPT',
    512: 'FLEX'
}
usage_data_with_projections['roster_slot_id'] = usage_data_with_projections['roster_slot_id'].map(lambda x: roster_slot_dict.get(x, x))

print(usage_data_with_projections['contest_type'].value_counts())
print(usage_data_with_projections['roster_slot_id'].value_counts())
usage_data_with_projections_filtered = usage_data_with_projections[usage_data_with_projections['contest_type'].isin(contest_types)].copy()
print(usage_data_with_projections_filtered['contest_type'].value_counts())
print(usage_data_with_projections_filtered['roster_slot_id'].value_counts())
# fix projections for captains
usage_data_with_projections_filtered.loc[usage_data_with_projections_filtered['roster_slot_id'] == 'CPT', 'projection'] *= 1.5
display(usage_data_with_projections_filtered)
usage_data_with_projections_filtered.to_csv('data/combined_data.csv')
```
